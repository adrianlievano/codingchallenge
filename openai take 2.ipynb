```python
def getwords(data):
    words=[]
    for i in range(len(data)):
        word=data[i]
        words.append(word)
    return words

def getletters(words):
    
    #Initialize chars and Letters matrix
    chars = []    
    fls=[]
    for x in range(len(words)):
        listChar=list(words[x])
        chars.append(listChar)   
    #return chars

    for i in range(len(words)):
        fl=chars[i][0:]
        fls.append(fl)
    return fls

def word_filter(data, letters, remove_word_size):
    remove = []
    if remove_word_size == 0:
        []
    else:
        for i in range(len(data)):
            if (len(letters[i][0:]) < remove_word_size+1 and len(letters[i][0:]) > remove_word_size-1):
                remove.append(data[i])
        return remove

def remove_list_ele(mas_list,remove_list):
    from collections import Counter
    a = mas_list
    b= remove_list
    c=list(Counter(a)-Counter(b))
    return c

def checker(output,data, prefix_level):
    data=sorted(data)
    if len(output) == len(data):
        flag = True
        print('Congrats! We identified all users using {} set of prefixes'.format(prefix_level))
    else:
        flag = False
        print('Find higher order features')
    return flag
#def check
```

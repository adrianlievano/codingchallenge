```python
#These functions prepare the dataset and format it for prefix-extraction algorithm

def getwords(data):
    '''This gets the string elements in a dataset'''
    words=[]
    for i in range(len(data)):
        word=data[i]
        words.append(word)
    return words

def getletters(words):
    '''This function takes an input of strings and converts them into a list of lists with each 
    character string of the original input'''
    
    #Initialize chars and Letters matrix
    chars = []    
    fls=[]
    for x in range(len(words)):
        listChar=list(words[x])
        chars.append(listChar)   
    #return chars

    for i in range(len(words)):
        fl=chars[i][0:]
        fls.append(fl)
    return fls

def word_filter(data, letters, remove_word_size):
    '''This function receives a dataset as a list, the letters of the words in the list, and a remove_word_size
    which specifies the length of word that should be filtered from the dataset'''
    remove = []
    if remove_word_size == 0:
        []
    else:
        for i in range(len(data)):
            if (len(letters[i][0:]) < remove_word_size+1 and len(letters[i][0:]) > remove_word_size-1):
                remove.append(data[i])
        return remove

def remove_list_ele(mas_list,remove_list):
    '''This function removes elements from two lists'''
    from collections import Counter
    a = mas_list
    b= remove_list
    c=list(Counter(a)-Counter(b))
    return c


def checker(output,data, prefix_level):
    '''This function compares the output data, which is iteratively rebuilt
    by prefix extraction algorithm, to the original database. When the data is completed extracted, the code ends'''
    data=sorted(data)
    if len(output) == len(data):
        flag = True
        print('Congrats! We identified all users using {} set of prefixes'.format(prefix_level))
    else:
        flag = False
        print('Find higher order features')
    return flag
```
